```python
data_df['Destination'].unique()
#data_df['PassengerId'].unique()
data_df['HomePlanet'].unique()
```

```python
data_df['Transported']=data_df['Transported'].astype(int)
data_df['CryoSleep'].replace({'True':1,'False':0},inplace=True)
data_df['VIP'].replace({'True':1,'False':0},inplace=True)
```

```python
dummi1=pd.get_dummies(data_df['CryoSleep'],drop_first=True,prefix='Cryo',)
dummi2=pd.get_dummies(data_df['VIP'],drop_first=True,prefix='VIP')
dummi3=pd.get_dummies(data_df['Destination'],prefix='Dest')
dummi4=pd.get_dummies(data_df['HomePlanet'],prefix='Home')
n_df=pd.concat([dummi1,dummi2,dummi3,dummi4],axis=1)
n_df
```

```python
types=data_df.columns.to_series().groupby(data_df.dtypes).groups
ctext=types[np.dtype('object')]
Columnas=data_df.columns
cnum=list(set(Columnas)-set(ctext))
print("Numericas= ",cnum)
data_ok=pd.concat([data_df[cnum],n_df],axis=1)
data_ok
```

### Matrix Correlacion

```python
corr = round(data_ok.corr(),3)
corr.style.background_gradient()
```
