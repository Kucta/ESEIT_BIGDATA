```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Diagrama de Dispersion (scatterplot)

x=np.arange(2,20)

y=x**2

plt.scatter(x,y, color="black", marker="*")
plt.xlabel("Variable x",fontsize=16, color="red")
plt.ylabel("Variable y", fontsize=16, color="red")
plt.title("Funcion de x, f(x)=x^2", fontsize=20, color="red")
plt.grid()
```

```python
#Diagrama de Dispersion (scatterplot)
# son buenos para analizar comportamiento entre variable

x=np.arange(2,20)
y2=x**3
y=x**2

plt.scatter(x,y, color="black", marker="*",label="x^2")
plt.scatter(x,y2, color="green", marker="+",label="x^3")
plt.plot(x,y2, color="pink", label="line")
plt.xlabel("Variable x",fontsize=16, color="red")
plt.ylabel("Variable y", fontsize=16, color="red")
plt.title("Funciones Polinomicas", fontsize=20, color="red")
plt.text(5,4000,"Texto",color="red")
plt.grid()
plt.legend()
```

```python
#Histogramas: para variables numericas unidimensionales

x=np.random.normal(10) #numero aleatorio alrededor de 10
mean =5
std=3
n=10000
x=np.random.normal(mean,std,n)
x
plt.title("Histograma de una Variable ALeatoria")
plt.ylabel("Frecuencia absoluta")
plt.xlabel("Variable")
plt.hist(x, bins=20);
plt.grid()
```

```python
plt.boxplot(x)
```

```python
# Subplots

fig, ax =plt.subplots(ncols=1,nrows=2,figsize=(5,10),sharex=True)
ax[0].hist(x);
ax[1].boxplot(x, vert=False);
```

```python
#BARPLOT

classes= ["gatos","perror","ratones"]
totales=[20,25,12]

plt.barh(classes,totales)
```

```python
import seaborn as sns

data=pd.DataFrame({
    "":classes,
    "totales":totales
})

data
plots=sns.barplot(x="animales",y="totales",data=data);

for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '1.f'),
    (bar.get_x()+bar.get_y()))
```
