### VSIUALIZACION DE DATOS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Diagrama de Dispersion (scatterplot)

x=np.arange(2,20)

y=x**2

plt.scatter(x,y, color="black", marker="*")
plt.xlabel("Variable x",fontsize=16, color="red")
plt.ylabel("Variable y", fontsize=16, color="red")
plt.title("Funcion de x, f(x)=x^2", fontsize=20, color="red")
plt.grid()
```

```python
#Diagrama de Dispersion (scatterplot)
# son buenos para analizar comportamiento entre variable

x=np.arange(2,20)
y2=x**3
y=x**2

plt.scatter(x,y, color="black", marker="*",label="x^2")
plt.scatter(x,y2, color="green", marker="+",label="x^3")
plt.plot(x,y2, color="pink", label="line")
plt.xlabel("Variable x",fontsize=16, color="red")
plt.ylabel("Variable y", fontsize=16, color="red")
plt.title("Funciones Polinomicas", fontsize=20, color="red")
plt.text(5,4000,"Texto",color="red")
plt.grid()
plt.legend()
```
